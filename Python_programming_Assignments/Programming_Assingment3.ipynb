```python
### Identify the number is positive,negative or zero
a=int(input("Enter the number "))
if a < 0:
    print("the number is negative.")
elif a >0:
    print("the number is positive")
else:
    print("the number is zero")
```

```python
### check the number whether the number is odd or even
a=int(input("Enter the number "))
if (a%2)== 0:
    print("the number is even")
else:
    print("the number is odd")
```

```python
### check whether the year is leap year or not.
a=int(input("Enter the year "))
if(a%4 ==0):
    print("the given year is a leap year.")
else:
    print("the given year isnot a leap year")
    
        
```

```python
### check whether the number is prime number or not
a=int(input("Enter the number "))
status=0
if a > 1:
    for i in range(2,a):
        if (a%i)==0:
            status=1
            break
    if (status == 0):
        print("true")
    else:
        print("false")
else:
    print("false")
            
```
